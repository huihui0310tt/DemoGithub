```python
# Final Project : CPBL Champion - Pattern Mining
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
```

```python
# Read File

raw_data_cpbl_s0 = pd.read_excel('cpbl-all.xlsx', index_col=0, sheet_name = 0) 
raw_data_cpbl_s0 = raw_data_cpbl_s0.replace(np.nan, False)
raw_data_cpbl_s0 = raw_data_cpbl_s0.replace('Y', True)

raw_data_cpbl_s1 = pd.read_excel('cpbl-all.xlsx', index_col=0, sheet_name = 1)  # sheet_name(1) = preprocessing version
raw_data_cpbl_s1 = raw_data_cpbl_s1.replace(np.nan, False)
raw_data_cpbl_s1 = raw_data_cpbl_s1.replace('Y', True)

raw_data_cpbl_s0
```

```python
# Part1 : 對戰組合Mining
part1 = raw_data_cpbl_s0[['Uni-Lions 1st', 'Rakuten 1st', 'Fubon 1st', 'Brothers 1st', 'Dragons 1st', 'Whale 1st', 'Tiger 1st', 'Hank 1st', 'T-REX 1st', 
                       'Uni-Lions 2nd', 'Rakuten 2nd', 'Fubon 2nd', 'Brothers 2nd', 'Dragons 2nd', 'Whale 2nd', 'Tiger 2nd', 'Hank 2nd', 'T-REX 2nd']]
min_support = 0.03125   # 0.03125 = 1/32
 

frequent_1 = apriori(part1, min_support=min_support, use_colnames=True)
frequent_1 = frequent_1.sort_values('support', ascending=False)
frequent_1['total'] = frequent_1['support'] / 0.03125
frequent_1
```

```python
def table1query(table, query1, query2=None):
    
    if query2 is None:
        print('Now query : ' + query1 )
        temp = table[table[query1]]
        print(temp.index)
    else:
        print('Now query : '+ query1 + ' + '+ query2 )
        temp = table[table[query1] & table[query2]]
        print(temp.index)
    print()
```

```python
# Part1 Table Query
table1query(part1, 'Uni-Lions 1st')
table1query(part1, 'Brothers 1st')
table1query(part1, 'Rakuten 1st')
table1query(part1, 'Dragons 1st')
table1query(part1, 'Fubon 1st')

table1query(part1, 'Brothers 2nd')
table1query(part1, 'Uni-Lions 2nd')
table1query(part1, 'Fubon 2nd')
table1query(part1, 'Whale 2nd')
table1query(part1, 'Dragons 2nd')
table1query(part1, 'Rakuten 2nd')
table1query(part1, 'T-REX 2nd')
table1query(part1, 'Hank 2nd')
table1query(part1, 'Tiger 2nd')

table1query(part1, 'Rakuten 1st', 'Brothers 2nd')
table1query(part1, 'Brothers 1st', 'Uni-Lions 2nd')
table1query(part1, 'Rakuten 1st', 'Uni-Lions 2nd')
table1query(part1, 'Uni-Lions 1st', 'Brothers 2nd')
table1query(part1, 'Brothers 1st', 'Fubon 2nd')
table1query(part1, 'Uni-Lions 1st', 'Dragons 2nd')
table1query(part1, 'Uni-Lions 1st', 'Fubon 2nd')
table1query(part1, 'Uni-Lions 1st', 'Rakuten 2nd')
table1query(part1, 'Fubon 1st', 'Uni-Lions 2nd')
table1query(part1, 'Fubon 1st', 'Brothers 2nd')
table1query(part1, 'Fubon 1st', 'T-REX 2nd')
table1query(part1, 'Brothers 1st', 'Whale 2nd')
table1query(part1, 'Dragons 1st', 'Fubon 2nd')
table1query(part1, 'Dragons 1st', 'Whale 2nd')
table1query(part1, 'Dragons 1st', 'Tiger 2nd')
table1query(part1, 'Dragons 1st', 'Hank 2nd')
```

```python
# Part1 : 對戰組合Mining (Last 10 Years)
min_support = 0.1

frequent_1_tail10 = apriori(part1.tail(10), min_support=min_support, use_colnames=True)
frequent_1_tail10 = frequent_1_tail10.sort_values('support', ascending=False)
frequent_1_tail10['total'] = frequent_1_tail10['support'] / 0.1
frequent_1_tail10
```

```python
def table2query(table, query1, query2=None, special=None, b_all=None):
    print('Now query : ' + query1 , end = ', \t\t')
    
    a = len(table[table[query1]])
    
    if special is not None:
        b = len(table[table[query2] | table[special]])
    elif b_all is not None:
        b = len(table)
    else:
        b = len(table[table[query2]])

    print('Rate is : '+ str(a/b))
```

```python
# Part2 Table Query
table2query(part2, query1='G1W', query2='G1W', special='G1L')
table2query(part2, query1='G2W', query2='G2W', special='G2L')
table2query(part2, query1='G3W', query2='G3W', special='G3L')
table2query(part2, query1='G4W', query2='G4W', special='G4L')
table2query(part2, query1='G5W', query2='G5W', special='G5L')
table2query(part2, query1='G6W', query2='G6W', special='G6L')
table2query(part2, query1='G7W', query2='G7W', special='G7L')
table2query(part2, query1='G1L', query2='G1W', special='G1L')
table2query(part2, query1='G2L', query2='G2W', special='G2L')
table2query(part2, query1='G3L', query2='G3W', special='G3L')
table2query(part2, query1='G4L', query2='G4W', special='G4L')
table2query(part2, query1='G5L', query2='G5W', special='G5L')
table2query(part2, query1='G6L', query2='G6W', special='G6L')
table2query(part2, query1='G7L', query2='G7W', special='G7L')

table2query(part2, query1='F.H.C', b_all=True)
table2query(part2, query1='S.H.C', b_all=True)
table2query(part2, query1='A.S.C', b_all=True)

table2query(part2, query1='Pitcher_W', b_all=True)
table2query(part2, query1='Pitcher_SO', b_all=True)
table2query(part2, query1='Pitcher_ERA', b_all=True)
table2query(part2, query1='Pitcher_SV', b_all=True)
table2query(part2, query1='Pitcher_H', b_all=True)
table2query(part2, query1='Batter_H', b_all=True)
table2query(part2, query1='Batter_AVG', b_all=True)
table2query(part2, query1='Batter_RBI', b_all=True)
table2query(part2, query1='Batter_SB', b_all=True)
table2query(part2, query1='Batter_HR', b_all=True)

table2query(part2, query1='MVP_Pitcher', b_all=True)
table2query(part2, query1='MVP_Batter', b_all=True)
```
